## Domain Analysis

### **Domain Analysis of Heart Disease Prediction Features**  

**1. slope_of_peak_exercise_st_segment (int)**
   - Domain Significance:Indicates the slope of the ST segment during peak exercise in an ECG test.  
   - Impact:A downsloping ST segment is associated with ischemia (poor blood flow), increasing heart disease risk.  

**2. thal (categorical: normal, fixed_defect, reversible_defect)**  
   - Domain Significance:Results from a Thallium stress test, which evaluates blood flow to the heart.  
   - Impact: 
     - Normal: Healthy blood flow.  
     - Fixed Defect:Permanent damage (e.g., scar tissue from a heart attack).  
     - Reversible Defect:Temporary blockage, indicating ischemia.  

**3. resting_blood_pressure (int)**
   - Domain Significance:Measures blood pressure when the patient is at rest.  
   - Impact: 
     - High blood pressure (>130 mmHg) increases heart disease risk.  
     - Low blood pressure may indicate poor circulation.
       
**4. chest_pain_type (int, 4 categories)**
   - Domain Significance:Identifies different types of chest pain.  
   - Impact:  
     - Typical Angina:Likely due to coronary artery disease.  
     - Atypical Angina:Unusual pain, possibly heart-related.  
     - Non-Anginal Pain:Less likely to be heart-related.  
     - Asymptomatic:No pain, but heart disease might be present.  

5. **num_major_vessels (int, 0-3)** 
   - Domain Significance:The number of major coronary arteries visible in fluoroscopy.  
   - Impact:More blocked vessels indicate severe heart disease.  

6. **fasting_blood_sugar_gt_120_mg_per_dl (binary)** 
   - Domain Significance:Indicates whether fasting blood sugar exceeds 120 mg/dl.  
   - Impact:High levels are linked to diabetes, a risk factor for heart disease.  

7. **resting_ekg_results (int, values 0,1,2)**
   - Domain Significance:Electrocardiogram (ECG) findings at rest.  
   - Impact:Abnormal results may indicate arrhythmia or heart stress.

8. **serum_cholesterol_mg_per_dl (int)**
   -Domain Significance:Measures cholesterol levels in the blood.  
   - Impact:High cholesterol (>200 mg/dl) increases the risk of artery blockage.  

9. **oldpeak_eq_st_depression (float)**
   - Domain Significance:ST segment depression during exercise relative to rest.  
   - Impact:Higher depression indicates reduced blood supply to the heart (ischemia).  

10. **sex (binary: 0 = female, 1 = male)**
   - Domain Significance: Biological sex differences affect heart disease risk.  
   - Impact:Males generally have a higher risk of heart disease.
 
11. **age (int, in years)**
   - Domain Significance: Age is a key risk factor for heart disease.  
   - Impact:Risk increases significantly after 45 years (men) and 55 years (women).  

12. **max_heart_rate_achieved (int, BPM)**
   - Domain Significance:Measures peak heart rate during exercise.  
   - Impact:
     - Lower-than-expected max heart rate may indicate heart problems.  
     - Extremely high heart rate under stress might be risky.  

13. **exercise_induced_angina (binary: 0 = No, 1 = Yes)**
   - Domain Significance:Chest pain triggered by physical exertion.  
   - Impact:A strong indicator of reduced oxygen supply to the heart muscles.  

**Key Insights:**
- Features like chest pain type, ST segment slope, exercise-induced angina, and major vessel count are strong predictors of heart disease.  
- Blood pressure, cholesterol, blood sugar, and age contribute as risk factors.  
- ECG results and max heart rate indicate functional heart performance under stress.

## Removing outliers

```python
def remove_outliers_iqr(df, column):
    Q1 = new_df[column].quantile(0.25)
    Q3 = new_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return new_df[(new_df[column] >= lower_bound) & (new_df[column] <= upper_bound)]

# Remove outliers
df_cleaned = remove_outliers_iqr(new_df, 'serum_cholesterol_mg_per_dl')
df_cleaned = remove_outliers_iqr(df_cleaned, 'oldpeak_eq_st_depression')
```
